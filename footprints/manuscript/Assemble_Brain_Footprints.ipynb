```r
library(dplyr)
library(RPostgreSQL)
library(BiocParallel)
library(reshape2)
library(GenomicRanges)
library(purrr)
```

```r
# Define the footprint-reading function
# Create a function to read the files
readFootprintTable <- function(fp.path){
    tbl <- read.table(fp.path, sep = "\t", as.is = TRUE)
    colnames(tbl) <- c("chrom", "start", "end", "name", "score", "strand")
    return(tbl)
}
```

```r
# Read in all brain hint 20
brain.20.paths <- list.files("/ssd/mrichard/data/footprints/brain_hint_20/")
brain.20.paths <- file.path("/ssd/mrichard/data/footprints/brain_hint_20/",brain.20.paths)

register(MulticoreParam(workers = 30, stop.on.error = TRUE, log = TRUE), default = TRUE)

system.time(brain.20.list <- bptry(bplapply(brain.20.paths,
                              readFootprintTable
                              )))
```

```r
# Remove the weird "NA" column
register(MulticoreParam(workers = 30, stop.on.error = TRUE, log = TRUE), default = TRUE)
brain.20.list <- bplapply(brain.20.list, select, 1:6)
```

```r
# Combine the table and filter out only chr19
brain.20.list %>% bind_rows() %>% filter(chrom == "chr19") -> brain.20.df
```
